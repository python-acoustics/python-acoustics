```python
import numpy as np
import matplotlib.pyplot as plt
from acoustics import Signal
from acoustics.signal import Frequencies
```

```python
sample_rate = 44100
signal = Signal(np.random.randn(10000), sample_rate)
fig, ax = plt.subplots()
ax.plot(signal)
```

```python
s = Signal(signal, sample_rate)
```

```python
# dB in octave bands:
f, o = s.octaves()
Lp = o.leq()
fig, ax = plt.subplots()
ax.plot(f.center, Lp, ".-")
```

```python
# db in specified bands:
freq_bands = np.array([[100, 200], [200, 400], [800, 1600], [3200, 6400]])
frequencies = Frequencies(center=freq_bands.mean(axis=1), lower=freq_bands[:, 0], upper=freq_bands[:, 1])
frequencies
```

```python
f, o = s.bandpass_frequencies(frequencies=frequencies)
Lp = o.leq()
fig, ax = plt.subplots()
ax.plot(f.center, Lp, ".-")
```
